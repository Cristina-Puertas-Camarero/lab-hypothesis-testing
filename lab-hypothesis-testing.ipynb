```python
#libraries
import pandas as pd
import scipy.stats as st
import numpy as np

```

```python
df = pd.read_csv("https://raw.githubusercontent.com/data-bootcamp-v4/data/main/pokemon.csv")
df
```

- We posit that Pokemons of type Dragon have, on average, more HP stats than Grass. Choose the propper test and, with 5% significance, comment your findings.

```python
from scipy.stats import ttest_ind

# Filtrar los datos para obtener los HP stats de los tipos Dragon y Grass
dragon_hp = df[df['Type 1'] == 'Dragon']['HP']
grass_hp = df[df['Type 1'] == 'Grass']['HP']

# Realizar el test t de muestras independientes
t_stat, p_value = ttest_ind(dragon_hp, grass_hp, equal_var=False, alternative='greater')

print(f"Estadístico t: {t_stat}")
print(f"p-valor: {p_value}")

# Evaluar el resultado
if p_value < 0.05:
    print("Rechazamos la hipótesis nula. Los Dragon-type Pokémon tienen, en promedio, más HP que los Grass-type Pokémon.")
else:
    print("No podemos rechazar la hipótesis nula. No hay evidencia suficiente para afirmar que los Dragon-type Pokémon tienen, en promedio, más HP que los Grass-type Pokémon.")
```

- We posit that Legendary Pokemons have different stats (HP, Attack, Defense, Sp.Atk, Sp.Def, Speed) when comparing with Non-Legendary. Choose the propper test and, with 5% significance, comment your findings.


```python
# Filtrar los datos para obtener las estadísticas de los Pokémon Legendarios y No Legendarios
legendary_stats = df[df['Legendary'] == True]
non_legendary_stats = df[df['Legendary'] == False]

# Realizar el test t de muestras independientes para HP
t_stat_hp, p_value_hp = ttest_ind(legendary_stats['HP'], non_legendary_stats['HP'], equal_var=False)
print(f"Estadístico t para HP: {t_stat_hp}")
print(f"p-valor para HP: {p_value_hp}")

# Realizar el test t de muestras independientes para Attack
t_stat_attack, p_value_attack = ttest_ind(legendary_stats['Attack'], non_legendary_stats['Attack'], equal_var=False)
print(f"Estadístico t para Attack: {t_stat_attack}")
print(f"p-valor para Attack: {p_value_attack}")

# Realizar el test t de muestras independientes para Defense
t_stat_defense, p_value_defense = ttest_ind(legendary_stats['Defense'], non_legendary_stats['Defense'], equal_var=False)
print(f"Estadístico t para Defense: {t_stat_defense}")
print(f"p-valor para Defense: {p_value_defense}")

# Realizar el test t de muestras independientes para Sp.Atk
t_stat_spatk, p_value_spatk = ttest_ind(legendary_stats['Sp. Atk'], non_legendary_stats['Sp. Atk'], equal_var=False)
print(f"Estadístico t para Sp.Atk: {t_stat_spatk}")
print(f"p-valor para Sp.Atk: {p_value_spatk}")

# Realizar el test t de muestras independientes para Sp.Def
t_stat_spdef, p_value_spdef = ttest_ind(legendary_stats['Sp. Def'], non_legendary_stats['Sp. Def'], equal_var=False)
print(f"Estadístico t para Sp.Def: {t_stat_spdef}")
print(f"p-valor para Sp.Def: {p_value_spdef}")

# Realizar el test t de muestras independientes para Speed
t_stat_speed, p_value_speed = ttest_ind(legendary_stats['Speed'], non_legendary_stats['Speed'], equal_var=False)
print(f"Estadístico t para Speed: {t_stat_speed}")
print(f"p-valor para Speed: {p_value_speed}")
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/data-bootcamp-v4/data/main/california_housing.csv")
df.head()
```

**We posit that houses close to either a school or a hospital are more expensive.**

- School coordinates (-118, 34)
- Hospital coordinates (-122, 37)

We consider a house (neighborhood) to be close to a school or hospital if the distance is lower than 0.50.

Hint:
- Write a function to calculate euclidean distance from each house (neighborhood) to the school and to the hospital.
- Divide your dataset into houses close and far from either a hospital or school.
- Choose the propper test and, with 5% significance, comment your findings.
 

```python
# Definir las coordenadas
coordenadas_escuela = (-118, 34)
coordenadas_hospital = (-122, 37)

# Definir la función de distancia euclidiana
def distancia_euclidiana(lat, lon, coord):
    return np.sqrt((lat - coord[0])**2 + (lon - coord[1])**2)


distancias_escuela = []
distancias_hospital = []
cerca_de_escuela_o_hospital = []

for i in range(len(df)):
    lat = df.iloc[i]['latitude']
    lon = df.iloc[i]['longitude']
    
    distancia_escuela = distancia_euclidiana(lat, lon, coordenadas_escuela)
    distancia_hospital = distancia_euclidiana(lat, lon, coordenadas_hospital)
    
    distancias_escuela.append(distancia_escuela)
    distancias_hospital.append(distancia_hospital)
    
    if distancia_escuela < 0.50 or distancia_hospital < 0.50:
        cerca_de_escuela_o_hospital.append(True)
    else:
        cerca_de_escuela_o_hospital.append(False)

df['distancia_a_escuela'] = distancias_escuela
df['distancia_a_hospital'] = distancias_hospital
df['cerca_de_escuela_o_hospital'] = cerca_de_escuela_o_hospital


```

```python
df.head()
```

```python
# Dividir el conjunto de datos
casas_cercanas = df[df['cerca_de_escuela_o_hospital']]
casas_lejanas = df[df['cerca_de_escuela_o_hospital']]
```

```python
from scipy.stats import ttest_ind

# Realizar la prueba t de muestras independientes
t_stat, p_value = ttest_ind(casas_cercanas['median_house_value'], casas_lejanas['median_house_value'], equal_var=False)


# Evaluar el resultado
if p_value < 0.05:
    print("Rechazamos la hipótesis nula. Las casas cercanas a una escuela o un hospital son significativamente más caras.")
else:
    print("No podemos rechazar la hipótesis nula. No hay evidencia suficiente para afirmar que las casas cercanas a una escuela o un hospital son más caras.")
```
